```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.nn import functional as F
import math
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
import numpy as np
from scipy import linalg
from torch.utils.data import DataLoader

class ImprovedUNet(nn.Module):
    def __init__(self, in_channels=1):
        super(ImprovedUNet, self).__init__()
        
        # Encoder path
        self.enc1 = nn.Sequential(
            nn.Conv2d(in_channels+1, 128, 3, padding=1),
            nn.GroupNorm(8, 128),
            nn.SiLU()
        )
        
        self.pool1 = nn.MaxPool2d(2)
        
        self.enc2 = nn.Sequential(
            nn.Conv2d(128, 256, 3, padding=1),
            nn.GroupNorm(8, 256),
            nn.SiLU()
        )
        
        self.pool2 = nn.MaxPool2d(2)
        
        # Bridge
        self.bridge = nn.Sequential(
            nn.Conv2d(256, 512, 3, padding=1),
            nn.GroupNorm(8, 512),
            nn.SiLU(),
            nn.Conv2d(512, 512, 3, padding=1),
            nn.GroupNorm(8, 512),
            nn.SiLU()
        )
        
        # Decoder path
        self.upconv2 = nn.ConvTranspose2d(512, 256, 2, stride=2)
        self.dec2 = nn.Sequential(
            nn.Conv2d(512, 256, 3, padding=1),
            nn.GroupNorm(8, 256),
            nn.SiLU()
        )
        
        self.upconv1 = nn.ConvTranspose2d(256, 128, 2, stride=2)
        self.dec1 = nn.Sequential(
            nn.Conv2d(256, 128, 3, padding=1),
            nn.GroupNorm(8, 128),
            nn.SiLU(),
            nn.Conv2d(128, in_channels, 3, padding=1)
        )
        
        # Time embedding
        self.time_mlp = nn.Sequential(
            SinusoidalPositionEmbeddings(128),
            nn.Linear(128, 128),
            nn.SiLU()
        )
    
    def forward(self, x, t):
        # Time embedding
        time_emb = self.time_mlp(t)
        time_emb = time_emb.unsqueeze(-1).unsqueeze(-1)
        time_emb = time_emb.expand(-1, -1, x.shape[2], x.shape[3])
        time_emb = time_emb[:, :1, :, :]
        
        # Input with time embedding
        x = torch.cat([x, time_emb], dim=1)
        
        # Encoder
        e1 = self.enc1(x)
        e1_pool = self.pool1(e1)
        
        e2 = self.enc2(e1_pool)
        e2_pool = self.pool2(e2)
        
        # Bridge
        b = self.bridge(e2_pool)
        
        # Decoder with skip connections
        d2 = self.upconv2(b)
        d2 = torch.cat([d2, e2], dim=1)
        d2 = self.dec2(d2)
        
        d1 = self.upconv1(d2)
        d1 = torch.cat([d1, e1], dim=1)
        d1 = self.dec1(d1)
        
        return d1

class SinusoidalPositionEmbeddings(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.dim = dim

    def forward(self, time):
        device = time.device
        half_dim = self.dim // 2
        embeddings = math.log(10000) / (half_dim - 1)
        embeddings = torch.exp(torch.arange(half_dim, device=device) * -embeddings)
        embeddings = time[:, None] * embeddings[None, :]
        embeddings = torch.cat((embeddings.sin(), embeddings.cos()), dim=-1)
        return embeddings

class ImprovedDDPM:
    def __init__(self, model, n_steps=1000, device='cuda'):
        self.model = model
        self.n_steps = n_steps
        self.device = device
        
        # Improved noise schedule
        self.beta = torch.linspace(1e-4, 0.02, n_steps).to(device)
        self.alpha = 1. - self.beta
        self.alpha_bar = torch.cumprod(self.alpha, dim=0)
        
        # Precalculate variables for sampling
        self.sqrt_alpha_bar = torch.sqrt(self.alpha_bar)
        self.sqrt_one_minus_alpha_bar = torch.sqrt(1. - self.alpha_bar)
    
    def noise_images(self, x, t):
        noise = torch.randn_like(x)
        noised_x = (
            self.sqrt_alpha_bar[t].view(-1, 1, 1, 1) * x +
            self.sqrt_one_minus_alpha_bar[t].view(-1, 1, 1, 1) * noise
        )
        return noised_x, noise
    
    def train_step(self, x, optimizer):
        optimizer.zero_grad()
        
        t = torch.randint(0, self.n_steps, (x.shape[0],), device=self.device)
        noisy_images, noise = self.noise_images(x, t)
        noise_pred = self.model(noisy_images, t.float())
        
        mse_loss = F.mse_loss(noise_pred, noise)
        l1_loss = F.l1_loss(noise_pred, noise)
        loss = mse_loss + 0.1 * l1_loss
        
        loss.backward()
        torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)
        optimizer.step()
        
        return loss.item()
    
    def sample(self, n_samples, size=(1, 28, 28)):
        """Generate new images using the trained model"""
        self.model.eval()
        with torch.no_grad():
            x = torch.randn((n_samples, *size)).to(self.device)
            
            for t in reversed(range(self.n_steps)):
                t_batch = torch.full((n_samples,), t, device=self.device, dtype=torch.long)
                predicted_noise = self.model(x, t_batch.float())
                
                alpha = self.alpha[t]
                alpha_bar = self.alpha_bar[t]
                beta = self.beta[t]
                
                if t > 0:
                    noise = torch.randn_like(x)
                else:
                    noise = torch.zeros_like(x)
                    
                x = 1 / torch.sqrt(alpha) * (x - ((1 - alpha) / torch.sqrt(1 - alpha_bar)) * predicted_noise) + torch.sqrt(beta) * noise
        
        return x

def calculate_activation_statistics(activations):
    mu = np.mean(activations, axis=0)
    sigma = np.cov(activations, rowvar=False)
    return mu, sigma

def calculate_fid(real_activations, generated_activations):
    mu1, sigma1 = calculate_activation_statistics(real_activations)
    mu2, sigma2 = calculate_activation_statistics(generated_activations)
    
    ssdiff = np.sum((mu1 - mu2) ** 2.0)
    covmean = linalg.sqrtm(sigma1.dot(sigma2))
    
    if np.iscomplexobj(covmean):
        covmean = covmean.real
    
    fid = ssdiff + np.trace(sigma1 + sigma2 - 2.0 * covmean)
    return fid

def evaluate_model(ddpm, test_loader, n_samples=1000):
    """Evaluate the DDPM model using various metrics"""
    print("Generating samples...")
    generated_samples = ddpm.sample(n_samples)
    
    print("Calculating MSE...")
    # Create a batch of real samples of the same size as generated samples
    real_samples = []
    for images, _ in test_loader:
        real_samples.append(images)
        if len(torch.cat(real_samples, dim=0)) >= n_samples:
            break
    real_samples = torch.cat(real_samples, dim=0)[:n_samples].to(ddpm.device)
    
    mse = torch.mean((generated_samples - real_samples) ** 2).item()
    
    print("Calculating FID...")
    real_features = real_samples.view(n_samples, -1).cpu().numpy()
    gen_features = generated_samples.view(n_samples, -1).cpu().numpy()
    fid = calculate_fid(real_features, gen_features)
    
    print("Generating visualization...")
    grid = make_grid(generated_samples[:16], nrow=4, normalize=True)
    plt.figure(figsize=(10, 10))
    plt.imshow(grid.cpu().permute(1, 2, 0))
    plt.axis('off')
    plt.show()
    
    return {
        'mse': mse,
        'fid': fid,
        'samples': generated_samples
    }

def train_and_evaluate_ddpm(device='cuda', n_epochs=100, batch_size=32):
    # Set up datasets
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))
    ])
    
    trainset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
    testset = torchvision.datasets.MNIST(root='./data', train=False, download=True, transform=transform)
    
    train_loader = DataLoader(trainset, batch_size=batch_size, shuffle=True, num_workers=2)
    test_loader = DataLoader(testset, batch_size=batch_size, shuffle=False, num_workers=2)
    
    # Initialize model and DDPM
    model = ImprovedUNet(in_channels=1).to(device)
    ddpm = ImprovedDDPM(model, device=device)
    
    # Setup training
    optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=0.01)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=n_epochs)
    
    # Training loop with periodic evaluation
    print("Starting training...")
    training_losses = []
    evaluation_metrics = []
    
    for epoch in range(n_epochs):
        model.train()
        epoch_losses = []
        
        for batch_idx, (x, _) in enumerate(train_loader):
            x = x.to(device)
            loss = ddpm.train_step(x, optimizer)
            epoch_losses.append(loss)
            
            if batch_idx % 100 == 0:
                print(f"Epoch {epoch+1}, Batch {batch_idx}, Loss: {loss:.4f}")
        
        avg_loss = sum(epoch_losses) / len(epoch_losses)
        training_losses.append(avg_loss)
        scheduler.step()
        
        # Evaluate every 10 epochs
        if (epoch + 1) % 10 == 0:
            print(f"\nEvaluating at epoch {epoch+1}...")
            metrics = evaluate_model(ddpm, test_loader)
            evaluation_metrics.append(metrics)
            print(f"MSE: {metrics['mse']:.4f}")
            print(f"FID: {metrics['fid']:.4f}\n")
        
        print(f"Epoch {epoch+1}/{n_epochs}, Average Loss: {avg_loss:.4f}")
    
    # Final evaluation
    print("\nFinal Evaluation:")
    final_metrics = evaluate_model(ddpm, test_loader)
    
    # Plot training progression
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(training_losses)
    plt.title('Training Loss Progression')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    
    plt.subplot(1, 2, 2)
    eval_epochs = list(range(9, n_epochs, 10))
    mse_scores = [m['mse'] for m in evaluation_metrics]
    fid_scores = [m['fid'] for m in evaluation_metrics]
    plt.plot(eval_epochs, mse_scores, label='MSE')
    plt.plot(eval_epochs, fid_scores, label='FID')
    plt.title('Evaluation Metrics Progression')
    plt.xlabel('Epoch')
    plt.ylabel('Score')
    plt.legend()
    plt.tight_layout()
    plt.show()
    
    return ddpm, training_losses, evaluation_metrics

# Run the complete training and evaluation
ddpm, losses, metrics = train_and_evaluate_ddpm()
```
